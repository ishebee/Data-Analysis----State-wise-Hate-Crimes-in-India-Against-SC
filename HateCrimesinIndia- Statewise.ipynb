```python
crime_by_state["Year"]= pd.to_datetime(crime_by_state["Year"], format='%Y').dt.year
years = crime_by_state["Year"].unique()
years
```

__attrocity_type__:- Using this variable just to take the list of attrocities

```python
attrocity_type = cols.tolist()
attrocity_type.remove('STATE/UT')
attrocity_type.remove('Year')
attrocity_type
```

__total_sum_crime_by_year__ :- Use this variable to create another data frame with total count for each attrocities per year.

__total_sum_crime_by_state__ :- Use this variable to create another data frame with total count for each attrocities per state.

__Total__ column is Dataframe to know count of the row

```python
total_sum_crime_by_year = crime_by_state.groupby("Year")[attrocity_type].sum()
total_sum_crime_by_year["Total"]=0
for attrocity in attrocity_type:
    total_sum_crime_by_year["Total"]= total_sum_crime_by_year["Total"]+total_sum_crime_by_year[attrocity]
total_sum_crime_by_year
```

```python
total_sum_crime_by_state = crime_by_state.groupby("STATE/UT")[attrocity_type].sum()
total_sum_crime_by_state["Total"]=0
for attrocity in attrocity_type:
    total_sum_crime_by_state["Total"]= total_sum_crime_by_state["Total"]+total_sum_crime_by_state[attrocity]
total_sum_crime_by_state
```

Let's plot bar graph for each state and total number of attrocities against Scheduled cast.

First of all we have to use only the required columns excluding STATE/UT & Year.

__Bar graph, Heat Map, Line Graph,Stacked Bar Graph__

```python
#Plotting Attrocity Graphs  with total count from 2001 - 2012 w.r.t STATE/UT 
for attrocity in attrocity_type:
    crime = crime_by_state.groupby("STATE/UT")[attrocity].sum()
    crime.plot.bar(color=['red', 'green', 'blue'],alpha=0.8)
    plt.xlabel('STATE/UT')
    plt.ylabel(attrocity + ' COUNT')
    plt.title('Bar Chart of Statewise ' + attrocity + ' Data')
    plt.show()
```

From the above bar graphs of attrocities w.r.t states in the country, we can find few states like Rajastan, Andhra Pradesh, Madhya Pradesh, Uttar Pradesh etc have high count of attrocites in all types.
